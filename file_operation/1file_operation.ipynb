#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

with open(")

```python

## read a whole file 

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## read a file line by line
# here r is the read mode
#  
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())##strip() helps to remove the newline character
```

```python
## writting a file(overwriting)
## here w is for write mode but it overwrites the content

with open('example.txt','w') as file:
    file.write("hello world\n")
    file.write("hello silchar\n")
```

```python
## writing a file(not overwriting)
with open('example.txt','a') as file:
    file.write("hello dibrugarh\n")
```

```python
## writting a list of lines to a file
line=['first line\n','second line\n','third line\n']
with open('example.txt','a') as file:
    file.writelines(line)
```

```python
## working with the binary files
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python

## reading a binary file

with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
## read the content from the source text file and copy the content and paste it to the destination text file
## copy the content 
with open('example.txt','r') as source_text:
    content=source_text.read()

## pasting it to the destination text file
with open('destination.txt','w') as destination_txt:
    destination_txt.write(content)
```

```python
#Read a text file and count the number of lines, words, and characters.
def count_text(file_path):
    with open(file_path,'r') as file:
## this will create a list =[first lline , second line , third line ]
        lines=file.readlines()
        line_count=len(lines)
        ## this will create [first line as ]=[ first, line ]
        word_count=  sum(  len(word.split())for word in lines)
        char_count= sum(  len(char)  for char in lines )
    return line_count,word_count,char_count


file_path='example.txt'
lines,word,char=count_text(file_path)
print(f'lines ; {lines}, word count :{word}, char count :{char}')
```
