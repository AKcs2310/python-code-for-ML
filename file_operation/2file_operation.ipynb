#### Working With File Paths
When working with files in Python, handling file paths correctly is crucial to ensure your code works across different operating systems and environments. Python provides several modules and functions for working with file paths effectively.

```python
## using the os model 

## creation a new directory
import os
new_directory="package"
os.mkdir(new_directory)
```

```python
## listing files and directories
## it returns all the files present in the directory in which we are working curerently

items=os.listdir('.')
print(items)
```

```python
## joining paths
directory_name='folder'
file_name='file.txt'
full_path=os.path.join(directory_name,file_name)
print(full_path)
```

```python

directory_name='folder'
file_name='file.txt'
## os.getcwd() it gives the current working directory

full_path=os.path.join(os.getcwd(),directory_name,file_name)
print(full_path)
```

```python
path='example1.txt'
if os.path.exists(path):
    print("the path do exist")
else :
    print("the path do not exist")
```

```python
## getting the absolute path
realtive_path='example.txt'
print(os.path.abspath(realtive_path))
```

```python
## both are same but the former is better 

full_path=os.path.join(os.getcwd(),directory_name,file_name)
print(full_path)
```
