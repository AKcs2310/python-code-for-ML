##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
tup=tuple()
lst=list()
print(type(lst))
print(type(tup))
```

```python
numbers=tuple([1,2,34,5,5]) ##tuple in the form of lists

print(numbers)
```

```python
numbers=list((1,2,3,4,5)) ##list in the form of tuple
print(numbers)
```

```python
mixed_tuple=(1,"hello world",3.14,True)
print(mixed_tuple)
```

```python
##accessing tuple elements
numbers=tuple([1,2,34,5,5])
numbers
```

```python
numbers[2]
```

```python
numbers[-1] ## accessing the last element
```

```python
numbers[0:4]
```

```python
numbers[::-1] ## printing the number in reverse order
```

```python
## tuple conacatetion
concatenation_tuple=numbers + mixed_tuple
```

```python
print(concatenation_tuple)
```

```python
mixed_tuple*3
```

```python
print(numbers)

numbers*3
```

```python
##immutable nature of tuples
##tuples are immutable in nature means that its elements cant be changed 
lst=[1,2,3,4]
print(lst)
lst[1]="krish"
print(lst)
```

```python
tup=tuple([1,2,3,4])
tup[0]=100
print(tup)
```

```python
numbers
```

```python
##tuples method
print(numbers.count(1))
print(numbers.index(34))
```

```python
## packing and unpacking of tuple
##packing
packed_tuple=1,"hello world ",3.14
print(packed_tuple)
```

```python
##unpacking of tuple
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with the *
numbers=tuple([1,2,3,4,5])
first,*middle,last=numbers
print(first)
print(*middle)
print(last)
```

```python
## nested list
lst=[[1,2,3,4],[6,7,8,9],[1,"hello",3.14,"c"]]
print(lst[0][0:4])
```

```python
print(lst[2][0:4])
```

```python
## nested tuple
nested_tuple=((1,2,3),(4,5,6),("hell0",3.10,88,"a"))
## accessing the element inside a tuple
print(nested_tuple[0])
```

```python
print(nested_tuple[2][0:5])
```

```python
## iterating over nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")

    print()
```

#### Conclusion
Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments and return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python code.
