##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(type(empty_dict))
```

```python
## creating a dict and always remember a single key is always used 
my_dict={"name":"ankit","age":22,"education":"btech","branch":"CSE","grade":7.03}
print(my_dict)
print(type(my_dict))
```

```python
## accessing Dictionary Elements
my_dict={"name":"ankit","age":22,"education":"btech","branch":"CSE","grade":7.03}
print(my_dict['name'])
print(my_dict["age"])
```

```python
## accessing dict using get method
print(my_dict.get('name'))
print(my_dict.get('age'))
print(my_dict.get('education'))
print(my_dict.get('branch'))
print(my_dict.get('grade'))
```

```python
## Modifying Dicitonary Elements
## Dictionary are mutable,so you can add, update or delete elements
## checking mutability
my_dict["name"]="ankit kumar gupta"
print(my_dict)
```

```python
## checking addition of new key value pair
my_dict["address"]="dibrugarh"
print(my_dict)
```

```python
## deleting a key value pair
del my_dict['name']
print(my_dict)
```

```python
## dictionaries method
## method to get al the keys from a dictionaries
all_keys=my_dict.keys()
print(all_keys)
## method to get all the values from a dictionaries
all_values=my_dict.values()
print(all_values)
## method to get all the items from a dictionaries
all_items=my_dict.items()
print(all_items)
```

```python
##  shallow copy
my_dict_copy=my_dict.copy()## use of copy function

print(my_dict)
print(my_dict_copy)
```

```python
my_dict["address"]="kolkata"
print(my_dict)
print(my_dict_copy)
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionatries, keys,values,or items

## Iterating over keys
for key in my_dict.keys():
    print(key)
```

```python
## iteration over values
for val in my_dict.values():
    print(val)
```

```python
## iterating over key value pair
for key,val in my_dict.items():
    print(f"{key}:{val}")
```

```python
## nested dictonaries
students={
    "student1":{"name":"ankit","age":22},
    "student2":{"name":"ajay","age":48},
    "student3":{"name":"neelam","age":44}


}
print(students)
```

```python
## accessing nested dictionaries


print(students["student1"]["name"])
```

```python
students.items()
```

```python
## iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,val in student_info.items():
        print(f"{key}:{val}")
```

```python
## dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dictionary comprehension
even_squares={x:x**2 for x in range(5) if x%2==0}
print(even_squares)
```

```python
## Practical Examples

## USe a dictionary to count the frequency of elements in list
numbers=[1,2,2,3,3,3,4,4,4,4]
freq={}
for num in numbers:
    if num in freq:
        freq[num]+=1
    else: freq[num]=1
print(freq)



```

```python
## merege two dictionaries
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
