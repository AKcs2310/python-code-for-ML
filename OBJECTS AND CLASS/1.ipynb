####  Classes and Objects
Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. This lesson covers the basics of creating classes and objects, including instance variables and methods.

```python

## instance variables and method
## defining a class with instance variable

class Dog:
    ## constructor

    def __init__(self,name,age):
        self.name=name
        self.age=age
```

```python
dog1=Dog("tommy",2)
```

```python
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
## defining a class with instance method
class Dog:
    ## constructor
    ## instance variable
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def bark(self):
        print(f"{self.name} says whoof!")
dog1=Dog("lucy",3)
dog1.bark()
dog2=Dog("tommy",4)
dog2.bark()
```

```python
## modelling a bank account
## create a class for bank account
class BankAccount:
    ## constructor
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
    def withdraw(self,moneywithdraw):
        if(moneywithdraw>self.balance):
            print("isufficient amount")
        else:
            self.balance-=moneywithdraw
            print(f"money withdraw is {moneywithdraw} successfully")

    def getbalance(self):
        print(f"the money present is {self.balance}")
```

```python
ankit=BankAccount("ankit",23)
ankit.getbalance()
ankit.deposit(23)
ankit.getbalance()
ankit.withdraw(88)
ankit.getbalance()
```
