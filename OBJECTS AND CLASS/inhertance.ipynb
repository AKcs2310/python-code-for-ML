#### Inheritance In Python
Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class. This lesson covers single inheritance and multiple inheritance, demonstrating how to create and use them in Python.

```python
## inheritance 
## parent class
## SINGLE INHERITANCE 

class Car:
    def __init__(self,window,door,enginetype):
        self.window=window
        self.door=door
        self.enginetype=enginetype
    def drive(self):
        print(f"the person gonna drive {self.enginetype} car!")
```

```python
audi=Car("black",3,"petrol")
audi.drive()
```

```python
##child class
class Tesla(Car):
    def __init__(self,window,door,enginetype,isself_driving):
        super().__init__(window,door,enginetype)
        self.isself_driving=isself_driving
    def selfdriving(self):
        print(f"the tesla car is self driving : {self.isself_driving}")
```

```python
tesla1=Tesla("black",3,"electric",True)
tesla1.selfdriving()
tesla1.drive()
```

```python
## multiple inheritance 
## inheriting from two or more base class
## base 1 
class Animal:
    def __init__(self, name):
        self.name=name
    def Anniename(self):
        return f" the name of the Animal is {self.name} "

## base 2 
    

class Pet:
    def __init__(self, owner):
        self.owner=owner

    def says(self):
        return f"the owner of the dog is {self.owner}"


## derived class from Animal and Pet
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)




dog1=Dog("tommy","ankit")
print(dog1.says())
print(dog1.Anniename())
```

#### Conclusion
Inheritance is a powerful feature in OOP that allows for code reuse and the creation of a more logical class structure. Single inheritance involves one base class, while multiple inheritance involves more than one base class. Understanding how to implement and use inheritance in Python will enable you to design more efficient and maintainable object-oriented programs.
