##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python



def even_or_odd(num):
    "function to find if the given number is ven or odd"
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## calling the function
even_or_odd(6)
```

```python
## function with multiple parameters
def add(a,b):
    return a+b
result=add(2,4)
print(result)
```

```python
## Default Parameters\
def greet(name):
    print(f"hello my name is {name}")
greet("ankit")
```

```python
def greet(name="Guest"):
    print(f"hello {name} how are you")
greet("ajay")
   
```

```python
### Variable Length Arguments
## Positional And Keywords arguments
## positional argument

def print_number(*args):
    for number in args:
        print(number)
```

```python
print_number(1,2,3,445,5,5,"hello")
```

```python
## keyword arguments
def print_key(**kwargs):
    for key,value in kwargs.items():


        print(f"{key} : {value}")
```

```python
print_key(name="ankit",age="22",country="India")
```

```python
##  mixture of positional and keyword arguments
def print_anything(*args,**kwargs):
    for num in args:
        print(f"positional argument {num}")
    for key,val in kwargs.items():
        print(f"{key}: {val}")
```

```python
print_anything(1,2,3,445,5,5,"hello",name="ankit",age="22",country="India")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
## return multiple parameters
def multiply(a,b):
    return a*b,b
```

```python
multiply(2,4)
```
