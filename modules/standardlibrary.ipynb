##### Standard Library Overview
Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr=array.array('i',[1,2,3])
print(arr.tolist())
##
##
##array.array() is a constructor that creates an array.

##The first argument 'i' specifies the type code, meaning the data type of elements in the array.
```

```python
import math
math.sqrt(16)
```

```python
import random
print(random.randint(1,10))
print(random.choice(['apple','banana']))
```

```python
## file and directory access
import os
print(os.getcwd())
os.mkdir('ankit.py')
```

```python
## High level operations on files and collection of files
import shutil
shutil.copy('source.txt','destination.txt')
```

```python
## Data Serialization
import json
data={'name':'ankit','age':22}
print(type(data))
json_str=json.dumps(data)
print(json_str)
print(type(json_str))
parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## csv file 
##ðŸ§© What is a CSV file?

##CSV stands for Comma-Separated Values.
##Itâ€™s a simple text file where each line represents a row of data, and each value is separated by a comma.

##Example of what example.csv might look like:
##name,age
##Krish,32
import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['ankit',22])
with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
from datetime import datetime,timedelta
now=datetime.now()
print(now)
yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
##regular expression
import re
pattern=r'\d+'
text="hi my name is 124 "
match=(re.search(pattern,text))
print(match.group())
```
